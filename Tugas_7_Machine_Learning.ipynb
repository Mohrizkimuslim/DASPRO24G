```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Load data
iris = load_iris()
X = iris.data

# 2. Scaling (Penting agar hasil akurat)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Menjalankan K-Means (K=3 sesuai jumlah spesies asli)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

# 4. Visualisasi
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, label='Centroids')

plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.title('K-Means Clustering (K=3)')
plt.legend()
plt.show()
```

```

# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris #dataset Iris
from sklearn.cluster import KMeans

# Membuat dataset Iris
iris = load_iris()
X = iris.data

# Menjalankan algoritma K-Means dengan jumlah cluster yang ditentukan
Kmeans = KMeans(n_clusters=6) #jumlah cluster
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Memvisualisasikan hasil clustering
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Menampilkan pusat cluster (centroid)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('K-Means Clustering (K=6)')
plt.show()
```
